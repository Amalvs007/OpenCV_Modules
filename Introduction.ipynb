```python
grey=cv.imread("E:\WALLPAPERS\896.jpg",cv.IMREAD_GRAYSCALE)
```

```python
cv.imshow("image",grey)
cv.waitKey()
cv.destroyAllWindows()
```

```python
# Converting to B/W image or Binary Image

(thresh,bw)=cv.threshold(grey,128,255,cv.THRESH_BINARY | cv.THRESH_OTSU)

# For coverting to black and white image, threshold needs to be given and 128 and 255 are the threshold values which adjust the blacksness and whiteness in the image.
# cv.THRESH_OTSU will automatically adjust the threshold values, which is 128 and 255.
# greay scale image is given as input as it is easy to convert grey scale image to balck and white image than an RGB image.
```

```python
cv.imshow("image",bw)
cv.waitKey()
cv.destroyAllWindows()
```

```python
# Edge Detection

edges=cv.Canny(grey,150,150)
cv.imshow("image",edges)
cv.waitKey()
cv.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=cv.imread("E:\WALLPAPERS\896.jpg")
plt.imshow(img)
```

```python
image=cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(image)

# OpenCv reads the image in BGR, but matplotlib reads the image in RGB.
# So, that BGR must be changed to RGB, for showing the image in matplotlib, otherwise it will be like the previous image.
```

```python
grey=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
plt.imshow(grey)
```

```python
img_copy=cv.rectangle(image,pt1=(200,100),pt2=(140,170),color=(0,0,255),thickness=3)
plt.imshow(img_copy)

# pt1 and pt2 are the coordinates of the image, where the rectangle should come.
```

```python
img_copy=cv.rectangle(image,pt1=(400,100),pt2=(140,170),color=(0,0,255),thickness=-3)
plt.imshow(img_copy)
```

```python
circleimg=cv.circle(image,center=(100,100),radius=30,color=(255,0,0),thickness=5)
plt.imshow(circleimg)
```

```python
text=cv.putText(image,text='Assasins Creed',org=(100,100),fontFace=cv.FONT_HERSHEY_DUPLEX,fontScale=7,color=(0,255,0),thickness=2)
plt.imshow(text)
```
