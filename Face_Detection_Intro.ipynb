```python
import cv2 as cv
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
face_cascade=cv.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
face_cascade
```

```python
img=cv.imread("E:\WALLPAPERS\A8S.jpg")
plt.imshow(img)
```

```python
image=cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
faces=face_cascade.detectMultiScale(image,1.1,4,minSize=(30,30))
```

```python
faces

# The array contains corrdinates of the image, where there is a face inside these coordinates.
```

```python
len(faces)

# This will give how many faces are detected in the image.
```

```python
for (x,y,w,h) in faces:
    cv.rectangle(image,(x-5,y-5),(x+w+5, y+h+5),(0,0,255),2)
```

```python
plt.imshow(image)
```

```python
cap=cv.VideoCapture("F:\Video Songs\One_Direction__Drag_Me_Down.mp4")
```

```python
while True:
    ret,frame = cap.read()
    grey = cv.cvtColor(frame,cv.COLOR_BGR2GRAY)
    faces=face_cascade.detectMultiScale(grey,1.1,4,minSize=(30,30))

    for(x,y,w,h) in faces:
        cv.rectangle(frame,(x,y),(x+w,y+h),(0,0,255),2)

    image = cv.resize(frame, (200,200))    # For reducing the size (Resolution), if the size of the video is too high.
    cv.imshow("video",image)               # If size is ok, then ignore the above code and type 'frame' instead of 'image' in this code.
    if cv.waitKey(1)& 0xFF ==27:
        break

cap.release()
cv.destroyAllWindows()
```

```python
# Detecting face from webcam

cap=cv.VideoCapture(0)
```

```python
while True:
    ret,frame = cap.read()
    grey = cv.cvtColor(frame,cv.COLOR_BGR2GRAY)
    faces=face_cascade.detectMultiScale(grey,1.1,4,minSize=(30,30))

    for(x,y,w,h) in faces:
        cv.rectangle(frame,(x,y),(x+w,y+h),(0,0,255),2)
   
    cv.imshow("video",frame)              
    if cv.waitKey(1)& 0xFF ==27:
        break

cap.release()
cv.destroyAllWindows()
```

```python
# Eye Detection

img = cv.imread("E:\WALLPAPERS\A8S.jpg")
```

```python
eye_cascade=cv.CascadeClassifier('haarcascade_eye.xml')
```

```python
eye=eye_cascade.detectMultiScale(img,1.1,4,minSize=(30,30))
```

```python
eye
```

```python
len(eye)
```

```python
for (x,y,w,h) in eye:
    cv.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
```

```python
# Eye Detection from webcam

cap = cv.VideoCapture(0)
```

```python
while True:
    ret,frame = cap.read()
    grey = cv.cvtColor(frame,cv.COLOR_BGR2GRAY)
    eyes=eye_cascade.detectMultiScale(grey,1.1,4,minSize=(30,30))

    for(x,y,w,h) in eyes:
        cv.rectangle(frame,(x,y),(x+w,y+h),(0,0,255),2)
   
    cv.imshow("video",frame)              
    if cv.waitKey(1)& 0xFF ==27:
        break

cap.release()
cv.destroyAllWindows()
```
