```python
import cv2
import numpy as np 
import mediapipe as mp 
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('hand.jpg')
```

```python
plt.imshow(img[:,:,::-1])

# Anothe way for getting the image in RGB.
```

```python
mp_hands = mp.solutions.hands

# This will detect whether, if there is hands in the image.
```

```python
hands = mp_hands.Hands(static_image_mode=True,max_num_hands=2,min_detection_confidence=0.3)

# For getting encodings of the detected Hands.
```

```python
mp_drawing=mp.solutions.drawing_utils

# For drawing the hands by connecting the coordinates.
```

```python
results=hands.process(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
# This will find the coordinates of the given image, also image must be changed to RGB.

if results.multi_hand_landmarks:
    for hand_no,hand_landmarks in enumerate(results.multi_hand_landmarks):
        mp_drawing.draw_landmarks(image=img,landmark_list=hand_landmarks,connections=mp_hands.HAND_CONNECTIONS)
```
