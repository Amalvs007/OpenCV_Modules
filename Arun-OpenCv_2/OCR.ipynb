## Image Processing

```python
img = cv.imread('Data/ocr_test.jpg')
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)

# Grey scale image
def get_greyScale(image):
    return cv.cvtColor(image,cv.COLOR_BGR2GRAY)

# Noise Removal image
def remove_noise(image):
    return cv.medianBlur(image,5)

# Grey scale image
def threshold(image):
    return cv.threshold(image,0,255,cv.THRESH_BINARY)[1]
```

```python
gray = get_greyScale(img)
```

```python
thresh = threshold(gray)
```

```python
noise = remove_noise(gray)
```

```python
plt.subplot(231)
plt.title("gray")
plt.imshow(gray)
plt.show()

plt.subplot(232)
plt.title("noise")
plt.imshow(noise)
plt.show()

plt.subplot(233)
plt.title("threshold")
plt.imshow(thresh)
plt.show()
```

```python
start = time.time()
print("Text extracted using Tesseract")
print(pytesseract.image_to_string(gray))
end = time.time()
total_time = end - start
print(f"Time taken {total_time} seconds")
```

```python
start = time.time()
print("Text extracted using Tesseract")
print(pytesseract.image_to_string(thresh))
end = time.time()
total_time = end - start
print(f"Time taken {total_time} seconds")
```

```python
start = time.time()
print("Text extracted using Tesseract")
print(pytesseract.image_to_string(noise))
end = time.time()
total_time = end - start
print(f"Time taken {total_time} seconds")
```
